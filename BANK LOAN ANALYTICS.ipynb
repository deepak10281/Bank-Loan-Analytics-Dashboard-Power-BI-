# KPI CALCULATIONS

```python
total_loan_applications = df['id'].count()

total_funded_amount = df['loan_amount'].sum()

total_amount_received = df['total_payment'].sum()

average_interest_rate = df['int_rate'].mean()

average_dti = df['dti'].mean()


print("Total Loan Applications :", total_loan_applications)

print("Total Funded Amount :", total_funded_amount)

print("Total Amount Received :", total_amount_received)

print("Average Interest Rate :", round(average_interest_rate,2))

print("Average DTI :", round(average_dti,2))
```

# GOOD LOAN ANALYSIS and BAD LOAN ANALYSIS

```python
good_loan = df[
    df['loan_status'] == 'Fully Paid'
]
```

```python
bad_loan = df[
    df['loan_status'] == 'Charged Off'
]
```

# GOOD LOAN % and BAD LOAN %

```python
# Good Loan %
good_loan_percentage = (
    len(good_loan)
    /
    len(df)
) * 100

print(round(good_loan_percentage,2))
```

```python
# Bad Loan %
bad_loan_percentage = (
    len(bad_loan)
    /
    len(df)
) * 100

print(round(bad_loan_percentage,2))
```

```python
labels = ['Good Loan', 'Bad Loan']

values = [
    good_loan_percentage,
    bad_loan_percentage
]

fig = go.Figure(
    data=[
        go.Pie(
            labels=labels,
            values=values,
            hole=0.7
        )
    ]
)

fig.show()
```

# MONTHLY LOAN TREND

```python
df['month'] = df['issue_date'].dt.month_name()

monthly_applications = df.groupby(
    'month'
)['id'].count()
```

```python
monthly_applications.plot(
    kind='line',
    marker='o'
)

plt.title("Monthly Loan Applications")

plt.ylabel("Applications")

plt.show()
```
